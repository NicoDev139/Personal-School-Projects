### Gas prices increase in different countries

```python
# reads gas_prices.csv
gas = pd.read_csv('gas_prices.csv')

# resizes for easy Readability 
# specifying number of pixels via dpi 
plt.figure(figsize=(8,5), dpi=300) 


# Displays Graph Title
plt.title('Gas Prices over Time (in USD)')



# Displays countries' gas_price increase
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.')
        


# prints tick-marks for every 3 years
print(gas.Year[: :3])



# Labels 'Y-axis' in 3 Year increaments 
plt.xticks(gas.Year[::3].tolist()+[2022])


# Adding Labels to 'X' and 'Y' axis
plt.xlabel('Year')
plt.ylabel('US Dollars')


# Applies Legend for each country
plt.legend(['Australia','Canada','France','Germany','Italy','Japan','Mexico','South Korea','UK', 'USA'])


# Save Graph
plt.savefig('Gas_price_figure1.png')

# Command to show graph. 
plt.show()
```

### Histogram to explore FIFA Dataset

#### Load Fifa Data

```python
# Dataset contains statistics on Soccer Players from-
# FIFA-18 Video Game
fifa = pd.read_csv('fifa_data.csv')

fifa
```

```python
# Displaying ony 10 rows of data
fifa.head(10)
```

### Histograms

```python
# Player's skill level intervals of tenths
bins = [40,50,60,70,80,90,100]

# Plot the overall skill level of each Player. 
plt.hist(fifa.Overall, bins=bins, color='#abddef')

# Labels the xticks per bins'
plt.xticks(bins)


# 'Y-label' gets number of players
plt.ylabel('Number of Players')

# 'X-label' gets skill level
plt.xlabel('Skill Level')

# Title of Histogram
plt.title('Distribution of Player Skills in FIFA 2019-Video Game')

# Save Graph
plt.savefig('Distribution_of_Player_Skills_FIFA.png')
plt.show()
```

### Pie Charts

```python
# Identify Players that play with either Left foot or Right foot
fifa.head(20)
```

```python
# Prints the Preferred Foot Column
fifa['Preferred Foot']
```

```python
# Count the number of people who play with their left foot
# use pandas' .loc to filter dataset by condition. i.e. 'left foot'
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
left
```

```python
# Count the number of people who play with their right foot
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
right
```

```python
# Display Pie Chart to show difference btwn left vs. right foot preference 
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

# Applying labels to each part of graphic
labels = ['Left', 'Right']

# Applies Two different colors to graph
colors = ['#abcdef', '#aabbcc']

# Command to display pie-chart
# displays labels, colors and percentages
plt.pie([left, right], labels = labels, colors = colors, autopct='%.2f %%')

# Adds title to Pie Chart
plt.title('Foot Preference of FIFA Players')

# Save Graph
plt.savefig('Foot_Preference_FIFA.png')

plt.show()
```

```python
# Condition applies to remove 'lbs'
# Allows to show numeric values only by converting it into 'int-value'
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

# Displays a full list of Player's Weight in numerical values
fifa.Weight
```

```python
fifa.Weight[0]
```

```python
# Counting Players under different weight categories

light = fifa.loc[fifa.Weight < 125].count()[0]

light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]

medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]

medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]

heavy = fifa[(fifa.Weight >= 200)].count()[0]

light
```

```python
# total of Players in light_medium category
light_medium
```

```python
# Total of Players in medium category
medium
```

```python
# Total of Players in medium_heavy category
medium_heavy
```

```python
# Total of Players in heavy category
heavy
```

```python
plt.style.use('ggplot')

weights = [light, light_medium, medium, medium_heavy, heavy] 

labels = ['Under 125', '125-150', '150-175','175-200','Over 200']
explode = (.4, .2, 0, 0, .4)

plt.title('Weight Distribution of FIFA Players (in lbs)')

# Save Graph
plt.savefig('Weigth_Category_FIFA.png')

plt.pie(weights, labels = labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.show()
```

### Box and Whisker Chart

```python
# Comparing how good teams are within dataset
# Comparing strenghts, Higher Score, Best Player, Worst Player
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
barcelona
```

```python
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
madrid
```

```python
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
revs
```

```python
# Passes all values from each team
plt.style.use('default')

plt.figure(figsize=(6,8))

# Compares Teams
# Shows the strength or weakness of each team
# Predicts both teams are strong contenders
labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']

boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    # Set edge color
    box.set(color= '#4286f4', linewidth=2)
    
    # Change Fill Color
    box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

# Save Graph
plt.savefig('Professional_Soccer_Team_Competition_FIFA.png')
plt.show
```
